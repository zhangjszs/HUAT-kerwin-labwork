**为了简便大家，整理为了一个bash脚本，可以直接运行，不需要手动输入。大家在当前目录创建一个`run.sh`文件，然后将下面的内容复制进去，然后运行`bash run.sh`即可。**
# 这是第一个

```bash
#!/bin/bash

# Step 1: Copy the Kafka tarball to /opt
cp /data/bigfiles/kafka_2.12-2.4.0.tgz /opt

# Step 2: Change to /opt directory and extract the tarball
cd /opt
tar -xzf kafka_2.12-2.4.0.tgz

# Step 3: Change into the Kafka directory
cd kafka_2.12-2.4.0

# Step 4: Start the Zookeeper service in the background
bin/zookeeper-server-start.sh config/zookeeper.properties &

# Step 5: Start the Kafka server in the background
bin/kafka-server-start.sh config/server.properties &

# Step 6: Wait for a moment to give Kafka time to start up
sleep 10

# Check if Kafka is running by trying to list topics
if bin/kafka-topics.sh --list --bootstrap-server localhost:9092 > /dev/null 2>&1; then
    echo "Kafka 启动成功。"
else
    echo "Kafka 启动失败，请检查错误日志。"
fi
```
# 这是第二个
```bash
#!/bin/bash

# Step 1: Copy the Kafka tarball to /opt
cp /data/bigfiles/kafka_2.12-2.4.0.tgz /opt

# Step 2: Change to /opt directory and extract the tarball
cd /opt
tar -xzf kafka_2.12-2.4.0.tgz

# Step 3: Change into the Kafka directory
cd kafka_2.12-2.4.0

# Step 4: Start the Zookeeper service in the background
bin/zookeeper-server-start.sh config/zookeeper.properties &

# Step 5: Start the Kafka server in the background
bin/kafka-server-start.sh config/server.properties &

# Step 6: Wait for a moment to give Kafka time to start up
sleep 10

# Step 7: Create a topic named 'topic_test' with 1 partition and 1 replication factor
bin/kafka-topics.sh --create --topic topic_test --partitions 1 --replication-factor 1 --bootstrap-server localhost:9092

# Step 8: Use Kafka producer to send messages to the topic
echo "I love Kafka" | bin/kafka-console-producer.sh --broker-list localhost:9092 --topic topic_test
echo "Kafka is good" | bin/kafka-console-producer.sh --broker-list localhost:9092 --topic topic_test

# Step 9: Confirm that the messages were sent successfully
echo "Messages sent successfully to topic_test."

# Optional: Verify the messages by consuming them
# bin/kafka-console-consumer.sh --bootstrap-server localhost:9092 --topic topic_test --from-beginning
```
