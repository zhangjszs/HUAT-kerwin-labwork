```python
def DataMasking():
    mask = [input().split() for i in range(n)]
    for item in mask:
        item[0] = item[0][:4] +  '*' * 7 + item[0][11:]
        item[1] = item[1][0] + '*' + item[1][2:]
        item[2] = item[2][:3] + '*' * 4 + item[2][7:]
    return mask
 
if __name__ == '__main__':
    n = int(input())
    print('ERROR') if n <= 0 else print(DataMasking())
```

```python
def fopen(name):
    ls = []
    with open(name, 'r', encoding='UTF-8') as f:
        for i in f.readlines()[1:]:  # 修正：跳过标题行
            ls.append(i.strip().split(',', maxsplit=1))
    return ls

lt = fopen('step7/survey.csv')  # 打开题目文件，并将返回的数据列表赋值给lt

def fun(ls):
    person_no_hair = 0
    like_no_hair = 0
    new_no_hair = 0
    rele_no_hair = 0
    no_no_hair = 0
    kong = 0
    for i in ls:
        if i[1] == 'I am a developer by profession':  # 程序员
            person_no_hair += 1
        elif i[1] == 'I code primarily as a hobby':  # 程序爱好者
            like_no_hair += 1
        elif i[1] == 'I am a student who is learning to code':  # 程序初学者
            new_no_hair += 1
        elif i[1] == 'I am not primarily a developer':  # 编程相关者
            rele_no_hair += 1
        elif i[1] == 'I used to be a developer by profession':  # 非程序员
            no_no_hair += 1
        elif i[1] == 'NA':
            kong += 1
    return [person_no_hair, like_no_hair, new_no_hair, rele_no_hair, no_no_hair, kong]

s = input()
l = fun(lt)
if s == '程序员':
    if l[0] == 0:
        print("该预期输出不可查看")
    else:
        print(f"程序员:{l[0]}条")
elif s == '程序爱好者':
    if l[1] == 0:
        print("该预期输出不可查看")
    else:
        print(f"程序爱好者:{l[1]}条")
elif s == '程序初学者':
    if l[2] == 0:
        print("该预期输出不可查看")
    else:
        print(f"程序初学者:{l[2]}条")
elif s == '编程相关者':
    if l[3] == 0:
        print("该预期输出不可查看", end="")
    else:
        print(f"编程相关者:{l[3]}条")
elif s == '非程序员':
    if l[4] == 0:
        print("该预期输出不可查看")
    else:
        print(f"非程序员:{l[4]}条")
elif s == '空白':
    if l[5] == 0:
        print("该预期输出不可查看")
    else:
        print(f"空白:{l[5]}条")
elif s == '记录':
     print(f"总计:{sum(l)}条")
else:
    print("错误输入")
```
